```python

def valores_nulos(df): 
    valores_nulos_por_columna = df.isnull().sum()
    valores_nulos_dataframe = df.isnull().sum().sum()
    return("Nulos por columna", valores_nulos_por_columna,
           "Nulos por dataframe", valores_nulos_dataframe)
```

```python
# Definir la función para cabmiar los valores nulos por la media de las columnas numéricas
def sustitucion_mean(df):
    # Seleccionar las columnas numéricas (int, float)
    columnas_numericas = df.select_dtypes(include=['float', 'int', 'float64', 'int64'])
    cualitativas = df.select_dtypes(include=["object","datetime","category"])
    # Reemplazar los valores nulos por la media de la columna
    numericas_mean = columnas_numericas.fillna(round(columnas_numericas.mean(), 1))
    # Unir data frame
    df = pd.concat([numericas_mean, cualitativas], axis=1)

    return df
```

```python
# Sustituir los valores nulos por la mediana de las columnas numéricas
def sustitucion_median(df):
    # Seleccionar las columnas 
    columnas_numericas = df.select_dtypes(include=['float', 'int', 'float64', 'int64'])
    cualitativas = df.select_dtypes(include=["object","datetime","category"])
    # Reemplazar los valores nulos por la media de la columna
    numericas_median = columnas_numericas.fillna(round(columnas_numericas.median(), 1))
    # Unir data frame
    df = pd.concat([numericas_median, cualitativas], axis=1)
    
    return df
```

```python
#Sustituir los valores nulos con el metodo ffill de las columnas object
def sustitucion_ffill(df):
    # Seleccionar las columnas
    columnas_numericas = df.select_dtypes(include=["float64","int64","float","int"])
    columnas_cualitativas = df.select_dtypes(include=["object","datetime","category"])
    # Reemplazar los valores nulos por el metodo ffill de la columna
    cualitativas = columnas_cualitativas.fillna(method="ffill")
    # Unir data frame
    df = pd.concat([cualitativas,columnas_numericas], axis=1)

    return df
```

```python
#Sustituir los valores nulos con el metodo bfill de las columnas object
def sustitucion_bfill(df):
    # Seleccionar las columnas
    columnas_numericas = df.select_dtypes(include=["float64","int64","float","int"])
    columnas_cualitativas = df.select_dtypes(include=["object","datetime","category"])
    # Reemplazar los valores nulos por el metodo bfill de la columna
    cualitativas = columnas_cualitativas.fillna(method="bfill")
    # Unir data frame
    df = pd.concat([cualitativas,columnas_numericas], axis=1)

    return df
```
