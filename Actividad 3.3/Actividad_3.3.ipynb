```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
archivo = input("escriba el nombre del archivo que quiere cargar: ")

def cargar_dataset(archivo):
    if '.csv' in archivo:
        df = pd.read_csv(archivo)
    elif '.xlsx' in archivo:
        df = pd.read_excel(archivo)
    elif '.json' in archivo:
        df = pd.read_json(archivo)
    elif '.html' in archivo:
        df = pd.read_html(archivo)
    return (df)
```

```python
df1 = cargar_dataset(archivo)
df1.to_csv("cargaCSV.csv")
df1
```

```python
df2 = cargar_dataset(archivo)
df2.to_csv("cargaXLSX.csv")
df2
```
